```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("obesity_data.csv")
df
```

```python
def plot_type(df, ageType, title, xlabel, ylabel, filename):
    plt.style.use('ggplot')
    x = np.array(df.columns[1:])
    y = df.loc[df["Type"]==ageType].to_numpy()[0][1:]
    print(x)
    print(y)
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.savefig(filename, bbox_inches='tight')
```

```python
plot_type(df, 'Adult', 'Obesity rate in adults vs Time', 'Time (year)', 'Obesity rate in adults in US', 'adults.png')
```

```python
plot_type(df, 'Adolescent', 'Obesity rate in adolescents vs Time', 'Time (year)', 'Obesity rate in adolescents in US', 'adolescents.png')
```
